```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
G = 6.67430e-11  # Gravitational constant (m^3 kg^-1 s^-2)
M = 1.989e30     # Mass of the Sun (kg)
m = 5.972e24     # Mass of the Earth (kg)
AU = 1.496e11    # Astronomical unit (m)

# Initial conditions (Earth's orbit)
# Position: 1 AU along x-axis, velocity in y-direction for circular orbit
initial_position = np.array([1.0 * AU, 0.0, 0.0])
initial_velocity = np.array([0.0, np.sqrt(G * M / AU), 0.0])

# Initial state vector [x, y, z, vx, vy, vz]
initial_state = np.concatenate((initial_position, initial_velocity))

def newtonian_gravity(t, state, G, M):
    """Compute the derivatives for the orbit problem."""
    x, y, z, vx, vy, vz = state
    
    # Calculate distance from central mass
    r = np.sqrt(x**2 + y**2 + z**2)
    
    # Acceleration due to gravity (F = ma = GMm/r^2 => a = GM/r^2)
    ax = -G * M * x / r**3
    ay = -G * M * y / r**3
    az = -G * M * z / r**3
    
    return [vx, vy, vz, ax, ay, az]

# Time span for simulation (1 year in seconds)
t_span = (0, 365.25 * 24 * 3600)
t_eval = np.linspace(t_span[0], t_span[1], 1000)

# Solve the ODE
solution = solve_ivp(newtonian_gravity, t_span, initial_state, 
                    args=(G, M), t_eval=t_eval, method='DOP853', 
                    rtol=1e-10, atol=1e-13)

# Extract the solution
position = solution.y[:3].T  # Transpose to get (N, 3) array
velocity = solution.y[3:].T

# Convert to AU for plotting
position_au = position / AU

# Create figure with subplots
fig = plt.figure(figsize=(15, 5))

# 3D Orbit Plot
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot(position_au[:, 0], position_au[:, 1], position_au[:, 2])
ax1.scatter([0], [0], [0], color='yellow', s=100, label='Sun')
ax1.set_xlabel('X (AU)')
ax1.set_ylabel('Y (AU)')
ax1.set_zlabel('Z (AU)')
ax1.set_title('3D Orbit')
ax1.legend()

# XY Projection
ax2 = fig.add_subplot(132)
ax2.plot(position_au[:, 0], position_au[:, 1])
ax2.scatter([0], [0], color='yellow', s=100, label='Sun')
ax2.set_xlabel('X (AU)')
ax2.set_ylabel('Y (AU)')
ax2.set_title('XY Plane Projection')
ax2.axis('equal')
ax2.legend()

# XZ Projection
ax3 = fig.add_subplot(133)
ax3.plot(position_au[:, 0], position_au[:, 2])
ax3.scatter([0], [0], color='yellow', s=100, label='Sun')
ax3.set_xlabel('X (AU)')
ax3.set_ylabel('Z (AU)')
ax3.set_title('XZ Plane Projection')
ax3.axis('equal')
ax3.legend()

plt.tight_layout()
plt.show()
```
